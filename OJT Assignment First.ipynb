# Q1. Data Visualization and Statistical Measures:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python


file_path =(r"C:\Users\RUTUJA\Downloads\iris (1).csv")
iris_data = pd.read_csv(file_path)
```

```python
iris_data.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8))
plt.show()
```

```python
#Box Plots for numericak columns
plt.figure(figsize=(10,8))
sns.boxplot(data=iris_data )
plt.title("Box Plots of Numerical Columns")
plt.show()
```

```python
def calculate_statistics(data):
    return {
        "Mean": data.mean(),
        "Median": data.median(),
        "Mode": data.mode().iloc[0],
        "Standard Deviation": data.std(),
        "Variance": data.var(),
        "Minimum": data.min(),
        "Maximum": data.max(),
        "Range": data.max() - data.min()
    }


numerical_data = iris_data.select_dtypes(include=['float64', 'int64'])


for column in numerical_data.columns:
    stats = calculate_statistics(numerical_data[column])
    print(f"Statistics for {column}:")
    for stat_name, value in stats.items():
        print(f"{stat_name}: {value}")
    print("\n")  
```
